*Calculate the mean daily return and daily standard deviation of returns, and then just annualise them to get mean expected annual return and volatility of that single stock.*

```
# Import pandas library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('/content/week4.csv')
df
```

*Mean daily return and daily standard deviation of returns:*

```
print('Mean daily return :',end=" ")
print(df['Day_Perc_Change'].mean()/100)
print('Daily standard deviation of returns :',end=" ")
print(df['Day_Perc_Change'].std()/100)
```

```
'''
annual mean = daily mean * 252 
annual stdev = daily stdev * sqrt(252)

'''
print('Annual Mean :',end=" ")
print(df['Day_Perc_Change'].mean()/100*252)
print('Annual standard deviation of returns :',end=" ")
print(df['Day_Perc_Change'].std()/100*(252**(0.5)))
```

```
df['Volatility'] = df['Day_Perc_Change'].rolling(window=252).std() * np.sqrt(252)
print(df['Volatility'])
```

####***5.3***

A scatter plot for differing weights of the individual stocks in the portfolio , the axes being the returns and volatility.  Colour the data points based on the Sharpe Ratio ( Returns/Volatility) of that particular portfolio.

```
df6
```

```
df6.columns
```

```
returns =df6[df6.columns[0:5]]

#calculate mean daily return and covariance of daily returns
mean_daily_returns = returns.mean()
cov_matrix = returns.cov()

#set number of runs of random portfolio weights
num_portfolios = 1000

#set up array to hold results
results = np.zeros((3,num_portfolios))
```

```
for i in range(num_portfolios):
    #select random weights for portfolio holdings
    w = np.random.random(5)
    #rebalance weights to sum to 1
    w /= np.sum(w)
    
    #calculate portfolio return and volatility
    portfolio_return = np.sum(mean_daily_returns * w) * 252
    portfolio_std_dev = np.sqrt(np.dot(w.T,np.dot(cov_matrix, w))) * np.sqrt(252)
    
    #store results in results array
    results[0,i] = portfolio_return
    results[1,i] = portfolio_std_dev
    #store Sharpe Ratio (return / volatility) - risk free rate element excluded for simplicity
    results[2,i] = results[0,i] / results[1,i]
#convert results array to Pandas DataFrame
results_frame = pd.DataFrame(results.T,columns=['ret','stdev','sharpe'])
#create scatter plot coloured by Sharpe Ratio
plt.ylabel('Returns')
plt.xlabel('Voltality')
plt.title('Sharpe Ratio ( Returns/Volatility)')
plt.scatter(results_frame.stdev,results_frame.ret,c=results_frame.sharpe,cmap='RdYlBu')
plt.colorbar()
```

```
#convert results array to Pandas DataFrame
results_frame = pd.DataFrame(results.T,columns=['ret','stdev','sharpe',df6.columns[0],df6.columns[1],df6.columns[2],df6.columns[3],df6.columns[4]])
```
