```
'''
Call should be buy whenever the smaller moving average (21) crosses over 
longer moving average (34) AND the call should be sell 
whenever smaller moving average crosses under longer moving average.
'''
print("enter the index of the day")
a=int(input())
k1=df7['PVR'].rolling(21).mean()[a]
k2=df7['PVR'].rolling(34).mean()[a]
if k1>k2:
    print("CALL IS :BUY")
else:
    print("CALL IS :SELL")
```

###***2.10***

**BOLLINGER BAND**

```
#The 14 day rolling mean of the closing price (we call it the average)
df['14 Day MA'] = df['Close Price'].rolling(window=13).mean()

df['14 Day STD'] = df['Close Price'].rolling(window=13).std()

#Upper band which is the rolling mean + 2 standard deviations away from the average
df['Upper Band'] = df['14 Day MA'] + (df['14 Day STD'] * 2)
#Lower band which is the rolling mean - 2 standard deviations away from the average.
df['Lower Band'] = df['14 Day MA'] - (df['14 Day STD'] * 2)
df.head(5)
```

```
df.tail()
```

```
fig = plt.figure(figsize=(12,6))
plt.grid(True)
plt.title('30 Day Bollinger Band for HUL')
plt.xlabel('DAYS')
plt.ylabel('Price')
plt.plot(df['Upper Band'],label= 'Upper Band')
plt.plot(df['Lower Band'],label= 'Lower Band')
plt.plot(df['14 Day MA'],label= 'MA 14 days')
plt.legend()
plt.show()
```
