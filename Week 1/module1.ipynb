####**1.5**

TO FIND AVERAGE PRICE AND PERCENTAGE LOSS/PROFIT :

```
'''
a function to calculate the average price over the last N days 
of the stock price data where N is a user defined parameter.
'''

def avg_price1(N):
  df2=df.tail(N)
  print(df2['Close Price'].mean())
```

```
'''
the average price AND the profit/loss percentages over the course of last 
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year
'''
print("AVERAGE PRICE OVER A WEEK IS ",end="")
avg_price1(7)  #week
print("AVERAGE PRICE OVER 2 WEEKS IS ",end="")
avg_price1(14)  # 2 weekS
print("AVERAGE PRICE OVER 1 MONTH IS ",end="")
avg_price1(30) 
print("AVERAGE PRICE OVER 3 MONTHS IS ",end="")
avg_price1(90) 
print("AVERAGE PRICE OVER 6 MONTHS IS ",end="")
avg_price1(150) 
print("AVERAGE PRICE OVER LAST YEAR IS ",end="")
avg_price1(365) 
```

```
# calculate the profit/loss percentage over the last N days.

def profit_loss(data1,n):
  data=data1.tail(n)
  a=0
  for i in range(0,n-1):
    change = data['Close Price'].iloc[i] - data['Close Price'].iloc[i+1]
    per  = (change*100)/data['Close Price'].iloc[i]
    a=a+per
  if(a>0):
    print('PROFIT %  IS ',end="")
    print(a)
  else:
    a=-a
    print('LOSS % IS ',end="")
    print(a)


print("FOR A WEEK  ",end="")
profit_loss(df,7)
print("FOR 2 WEEKS  ",end="")
profit_loss(df,14)
print("FOR A MONTH  ",end="")
profit_loss(df,30)
print("FOR 3 MONTHS  ",end="")
profit_loss(df,90)
print("FOR 6 MONTHS  ",end="")
profit_loss(df,150)
print("FOR 1 YEAR  ",end="")
profit_loss(df,365)
```

####**1.6**

CREATING A NEW COLUMN USING pct_change()

```
'''
column 'Day_Perc_Change' where the values are the daily change in percentages 
i.e. the percentage change between 2 consecutive day's closing prices.
'''
df['Day_Perc_Change']=(df['Close Price'].pct_change()*100)
df
```

```
df['Day_Perc_Change'].iloc[0]=0   #CHANGE NAN TO ZERO
df
```

####**1.7**

---



```
#trend column 
'''
Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive'  for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative'  for 'Day_Perc_Change' in between -0.5 and -1
'Positive'  for 'Day_Perc_Change' in between 1 and 3
'Negative'  for 'Day_Perc_Change' in between -1 and -3
'Among top gainers'  for 'Day_Perc_Change' in between 3 and 7
'Among top losers'  for 'Day_Perc_Change' in between -3 and -7
'Bull run'  for 'Day_Perc_Change'  &gt;7
'Bear drop'  for 'Day_Perc_Change'  &lt;-7

'''
df.loc[df['Day_Perc_Change'].between(-0.5,0.5),'Trend'] = 'Slight or No change'
df.loc[df['Day_Perc_Change'].between(0.5,1),'Trend'] = 'Slight positive'
df.loc[df['Day_Perc_Change'].between(-0.5,-1),'Trend'] = 'Slight negative'
df.loc[df['Day_Perc_Change'].between(1,3),'Trend'] = 'positive'
df.loc[df['Day_Perc_Change'].between(-1,-3),'Trend'] = 'negative'
df.loc[df['Day_Perc_Change'].between(3,7),'Trend'] = 'among top gainers'
df.loc[df['Day_Perc_Change'].between(-3,-7),'Trend'] = 'among top losers'
df.loc[df['Day_Perc_Change']>=7,'Trend'] = 'bull run'
df.loc[df['Day_Perc_Change']<=-7,'Trend'] = 'bear drop'
df['Trend'].fillna('Other', inplace=True)
```
